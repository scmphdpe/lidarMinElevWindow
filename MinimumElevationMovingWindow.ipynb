```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("selected_LIDAR_Points_3.txt")
```

```python
print(df.head())
print(len(df))
```

```python
west = min(df.POINT_X)
east = max(df.POINT_X)
south = min(df.POINT_Y)
north = max(df.POINT_Y)
```

```python
tl = (min(df.POINT_X), max(df.POINT_Y))
ew = east - west
ns = north - south
print(tl, ew, ns)
```

```python
window = 3.0
```

```python
df['index_X'] = np.trunc((df.POINT_X - west) / window)
df['index_Y'] = np.trunc((df.POINT_Y - south) / window)
```

```python
df.tail()
```

```python
df.astype({"index_X": int, "index_Y": int})
```

```python
pixels = []
for i in range(int(max(df.index_X))):
    for j in range(int(max(df.index_Y))):
        px = df.query(f'(index_X == {i}) & (index_Y == {j})')
        pixels.append([(np.around(x,6),
                        np.around(y,6),
                        np.around(z,6),
                        (i,j)) 
                        for x,y,z in zip(px.POINT_X,
                                         px.POINT_Y,
                                         px.POINT_Z)])
```

```python
print(len(pixels))
```

```python
pixNE = []
for x in range(len(pixels)):
    if pixels[x]:
        pixNE.append(pixels[x])
```

```python
print([str(x) for x in pixNE[:10]])
print(len(pixNE), type(pixNE))
```

```python
pixSet = [set(x) for x in pixNE]
```

```python
print(len(pixSet),pixNE[:5])
```

```python
with open('minElevLidar3.csv','w') as f:
    for i in pixSet:
        z = []
        for p in i:
            z.append(p[2])
        print(min(z))
        for q in i:
            if q[2] == min(z):
                outString = str(q[:3])[1:-1] + "\n"
                f.write(outString)
                #print(outString)
```
